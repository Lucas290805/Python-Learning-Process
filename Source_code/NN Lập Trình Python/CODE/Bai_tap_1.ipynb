Viết chương trình nhập vào số giờ làm mỗi tuần, thù lao trên mỗi giờ làm tiêu chuẩn, 
từ đó tính ra số tiền thực lĩnh của nhân viên. Với số giờ tiêu chuẩn mỗi tuần là 40 giờ, và 
mỗi giờ vượt chuẩn được trả gấp 1.5 lần so với giờ làm chuẩn.

```python
def tinh_luong():
    so_gio = float(input('nhap so gio lam viec: '))
    tien_moi_gio = float(input('tieng luong moi gio: '))

    gio_chuan = 40

    if so_gio == gio_chuan:
        tien_luong = so_gio * tien_moi_gio
    else:
        tien_luong_chuan = gio_chuan * tien_moi_gio
        gio_vuot = so_gio - gio_chuan
        tien_luong_vuot = gio_vuot * tien_moi_gio * 1.5
        tien_luong = tien_luong_vuot + tien_luong_chuan
        
    print(f'So tien luong cua nhan vien la: {tien_luong} VND')
    
tinh_luong()
```

```python
import math

def kiem_tra_chu_so_chan(n):
    
    #chuyển dữ liệu thành chuỗi để duyệt từng số
    chuoi_so = str(n)
    chuoi_so_chan = {'0', '2','4', '6','8'} 
    
    dem_so_chan = 0
    
    # nếu số trong chuỗi số là số chẵn thì tính, nếu có hơn hoặc bằng 2 thì là hợp lệ.
    for so in chuoi_so:
        if so in chuoi_so_chan:
            dem_so_chan += 1
    return dem_so_chan >=2

def tim_so_chinh_phuong(): # số chính phương là bình phương của 1 số
    
    #tạo khoảng để tìm x, vì căn 100 = 10
    min = 10
    max = int(math.sqrt(2000)) # xấp xỉ 45
    
    ket_qua = []
    
    for x in range(min, max +1): #vì tính cả 10 và 2000 nên phải +1
        so_chinh_phuong = x *x
        
        if kiem_tra_chu_so_chan(so_chinh_phuong):
            ket_qua.append(str(so_chinh_phuong))
            
    chuoi_ket_qua =  ','.join(ket_qua)
    print("Các số chính phương thỏa mãn điều kiện là:")
    print(chuoi_ket_qua)
    
tim_so_chinh_phuong()

            
```

```python
import re

def kiem_tra_username(username: str) ->bool:
#username yeu cau 6 ki tu
    if len(username) != 6:
        return False
# .isalum() dung de kiem tra neu khong phai chu va so, tra ve False    
    if not username.isalnum():
        return False   
# dkien khong duoc co chu in hoa
    if username != username.lower():
        return False   
    
    return True

#Định nghĩa mẫu Regex cho Password (Lookahead Assertions) Cú pháp: re.fullmatch(pattern, text)
#1. ^...$: Bắt đầu và kết thúc chuỗi (kiểm tra độ dài)
#2. (?=.*[a-z]): Phải có chữ thường
#3. (?=.*[A-Z]): Phải có chữ hoa
#4. (?=.*\d): Phải có số  (\d la viet tat cua [0,9])
#5. (?=.*[!@#$%^&*]): Phải có ký tự đặc biệt
#6. [a-zA-Z\d!@#$%^&*]{6,12}: Toàn bộ chuỗi chỉ chứa các ký tự cho phép và độ dài 6-12


PASSWORD_REGEX = r"^(?=.*[a-z])(?=.*[A-Z])(?=.*\d)(?=.*[!@#$%^&*])[a-zA-Z\d!@#$%^&*]{6,12}$"


def kiem_tra_password(password: str) -> bool:
    """Kiểm tra Password: 6-12 ký tự, phải có đủ 4 nhóm (thường, hoa, số, đặc biệt)."""
    
    # re.fullmatch đảm bảo toàn bộ chuỗi khớp với mẫu
    if re.fullmatch(PASSWORD_REGEX, password) is not None:
        return True
    else:
        return False

# --- Chương trình Chính ---

print("--- Đăng ký Tài khoản ---")
username_nhap = input("Nhập Username (6 ký tự, thường/số): ")
password_nhap = input("Nhập Password (6-12 ký tự, đủ 4 nhóm): ")

# Gọi các hàm kiểm tra
valid_user = kiem_tra_username(username_nhap)
valid_pass = kiem_tra_password(password_nhap)

# In kết quả
if valid_user and valid_pass:
    print("\n Đăng ký thành công! Username và Password hợp lệ.")
else:
    print("\n Đăng ký thất bại!")
    if not valid_user:
        print(f"   - Lỗi Username: '{username_nhap}' không hợp lệ (Phải đúng 6 ký tự, chỉ thường/số).")
    if not valid_pass:
        print(f"   - Lỗi Password: '{password_nhap}' không hợp lệ (Phải đủ 4 nhóm ký tự, 6-12 ký tự).")
```

Bài 15: Viết chương trình cho phép người dùng nhập vào một mảng hai chiều với mỗi phần
tử là một số nguyên, sau đó thực hiện các yêu cầu sau: - Tính tổng các số nguyên tố trong
mảng 2 chiều. - - Tính tổng đường chéo chính(trường hợp nhập vào mảng vuông) Tìm vị
trí của phần tử nhỏ nhất và lớn nhất trong mảng 2 chiều.

```python
import math

def la_so_nguyen_to(n):
    
#Một số tự nhiên (1, 2, 3, 4, 5, 6,...) được gọi là số nguyên tố nếu nó lớn hơn 1 và không thể được biểu diễn thành tích của hai số tự nhiên nhỏ hơn khác 1 vd: 7 = 2x3 +1
    if n < 2:
        return False

    for i in range (2, math.isqrt(n) + 1):
        if n % i == 0:
            return False
    return True

def nhap_mang(so_dong, so_cot):
    print(f"Nhập mảng: {so_dong}x{so_cot}:")
    matrix = []
    
    for i in range(so_dong):
# Dung map(a,b) de apply cac hanh dong a vao b.
        row = list(map(int, input(f'nhap dong: {i}').split()))
# Co the dung so luong phan tu trong 1 hang de kiem tra nguoi dung nhap du cot chua. vd: 1 2 3 = 3 cot       
        if len(row) != so_cot:
            print('Số lượng phần tử không khớp số cột!')
            continue
        matrix.append(row)
        break
    except ValueError:
        print("Lỗi: Chỉ được nhập số nguyên.")

def xu_ly_mang(matrix,so_dong, so_cot): 
    tong_so_nguyen = 0
    tong_duong_cheo = 0
    
    min_val = matrix[0],[0]
    min_pos = (0,0)
    
    max_val = matrix[0][0]
    max_pos = (0, 0)
    
    for i in range(so_dong):
        for j in range(so_cot):
            gia_tri = matrix[i],[j]
                   
#dkien 1: Tổng số nguyên tố
            if la_so_nguyen_to(gia_tri):
                tong_so_nguyen += gia_tri

#dkien 2:  Tổng đường chéo chính           
            if i == j
                tong_duong_cheo += gia_tri

#dkien 3:   Min/Max
            if gia_tri < min_val:
                min_val = gia_tri
                min_pos = (i, j)
            if gia_tri > max_val:
                min_pos = (i, j)

    return tong_so_nguyen, tong_duong_cheo, (min_val, min_pos), (max_val, max_pos)
```

```python
import math

def la_so_nguyen_to(n):
    """Kiểm tra số nguyên tố tối ưu."""
    if n < 2:
        return False
    for i in range(2, math.isqrt(n) + 1):
        if n % i == 0:
            return False
    return True

def main():
    # --- 1. Nhập kích thước mảng ---
    try:
        rows = int(input("Nhập số dòng: "))
        cols = int(input("Nhập số cột: "))
    except ValueError:
        print("Lỗi: Vui lòng nhập số nguyên.")
        return

    # --- 2. Nhập mảng 2 chiều ---
    print(f"Nhập các phần tử cho ma trận {rows}x{cols} (mỗi dòng nhập các số cách nhau bởi dấu cách):")
    matrix = []
    for i in range(rows):
        while True:
            try:
                # Nhập 1 dòng và ép kiểu sang list số nguyên
                row = list(map(int, input(f"-> Dòng {i}: ").split()))
                if len(row) != cols:
                    print(f" Lỗi: Bạn phải nhập đúng {cols} số. Hãy nhập lại dòng này.")
                    continue
                matrix.append(row)
                break
            except ValueError:
                print(" Lỗi: Chỉ được nhập số nguyên.")

    # --- 3. Xử lý logic ---
    tong_nt = 0
    tong_cheo_chinh = 0
    
    # Khởi tạo min/max tại phần tử đầu tiên
    min_info = {'val': matrix[0][0], 'pos': (0, 0)}
    max_info = {'val': matrix[0][0], 'pos': (0, 0)}

    for i in range(rows):
        for j in range(cols):
            val = matrix[i][j]
            
            # a. Tính tổng số nguyên tố
            if la_so_nguyen_to(val):
                tong_nt += val
            
            # b. Tính tổng đường chéo chính (chỉ cộng khi i == j)
            if i == j:
                tong_cheo_chinh += val

            # c. Tìm Min
            if val < min_info['val']:
                min_info['val'] = val
                min_info['pos'] = (i, j)
            
            # d. Tìm Max
            if val > max_info['val']:
                max_info['val'] = val
                max_info['pos'] = (i, j)

    # --- 4. Xuất kết quả ---
    print("\n" + "="*30)
    print(f"1. Tổng các số nguyên tố: {tong_nt}")

    if rows == cols:
        print(f"2. Tổng đường chéo chính: {tong_cheo_chinh}")
    else:
        print("2. Tổng đường chéo chính: Không tính được (Không phải mảng vuông)")

    print(f"3. Phần tử nhỏ nhất: {min_info['val']} tại vị trí dòng {min_info['pos'][0]}, cột {min_info['pos'][1]}")
    print(f"4. Phần tử lớn nhất: {max_info['val']} tại vị trí dòng {max_info['pos'][0]}, cột {max_info['pos'][1]}")
    print("="*30)

if __name__ == "__main__":
    main()
```
