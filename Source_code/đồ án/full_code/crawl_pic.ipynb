```python
import requests
import os
import time
from datetime import datetime

URL = "https://giaothong.hochiminhcity.gov.vn/render/ImageHandler.ashx?id=5a6065c58576340017d06615&t={timestamp}"
OUT_DIR = "C:/Users/HP/Videos/Captures/pic_not_label/row2"
INTERVAL_SECONDS = 15  # ví dụ: mỗi 5 giây lấy 1 ảnh

os.makedirs(OUT_DIR, exist_ok=True)

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)"
}

def download_image():
    ts = int(time.time() * 1000)
    url = URL.format(timestamp=ts)
    try:
        r = requests.get(url, headers=headers, timeout=10)
        r.raise_for_status()
        # kiểm tra kiểu content
        ct = r.headers.get("Content-Type", "")
        if not ct.startswith("image/"):
            print("Warning: không phải ảnh, content-type =", ct)
        fname = datetime.now().strftime("%Y%m%d_%H%M%S_%f")[:-3] + ".jpg"
        path = os.path.join(OUT_DIR, fname)
        with open(path, "wb") as f:
            f.write(r.content)
        print("Saved", path)
    except Exception as e:
        print("Error downloading:", e)

def main():
    try:
        while True:
            download_image()
            time.sleep(INTERVAL_SECONDS)
    except KeyboardInterrupt:
        print("Stopped by user")

if __name__ == "__main__":
    main()
```
